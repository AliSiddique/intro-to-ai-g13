# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load the Dataset

```python
df = pd.read_csv('../spotify_data.csv')
df.head()
```

# Look at the columns 

```python
df.columns
```

# Drop Unnamed: 0 , mode and key 
## Dropping these because they are not correlated to other features so they are not necessary for the machine learning models
### Unnamed: 0 is for the index, so we are replacing this with track_id
### Mode and key are not correlated with other features so we are dropping them

- [Dropfunction](https://stackoverflow.com/questions/13411544/delete-a-column-from-a-pandas-dataframe) - Drop a column

```python
df.drop(columns=['Unnamed: 0','mode','key'], inplace=True)

# Set 'track_id' as the new index
df.set_index('track_id', inplace=True)
df
```

# Find any null values in the dataset

```python
null_values = df.isnull()
null_values
```

# See how many null values there are

```python
null_counts = null_values.sum()
null_counts
```

```python
# There are over 1 million rows, we only want to keep 150,000 so we are taking the first 150,000 rows
```

```python
df = df.head(150000)
print(df.shape)
```

```python
df
```

# Visualising if the data distirbution has changed significantly which it hasn't
## We will use this dataset

```python
numeric_features = ['popularity', 'year', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']

# Set up subplots for histograms
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 10))
plt.subplots_adjust(hspace=0.5)

# Create histograms for each numeric feature
for i, feature in enumerate(numeric_features):
    row, col = divmod(i, 3)
    axes[row, col].hist(df[feature], bins=30, edgecolor='k', alpha=0.7)
    axes[row, col].set_title(f'Histogram of {feature}')
    axes[row, col].set_xlabel(feature)
    axes[row, col].set_ylabel('Frequency')

# Adjust the layout
plt.tight_layout()

# Show the histograms
plt.show()
```
